```python
import pandas as pd
import plotly.express as px

# Step 1: Read the CSV file
a = "F:/n/2d_front_view/000000.csv"
data = pd.read_csv(a)

# Step 2: Create the scatter plot using Plotly
fig = px.scatter(
    data, 
    x='Xfront', 
    y='Yfront', 
    color='intensity',  # Color by intensity values
    color_continuous_scale='gray',  # Use gray scale for intensity
    labels={'Xfront': 'Xfront', 'Yfront': 'Yfront'},  # Axis labels
    title='Scatter Plot of Xfront vs Yfront with Intensity'
)

# Step 3: Ensure equal scaling of both axes (aspect ratio is 1:1)
fig.update_layout(
    autosize=True,
    xaxis=dict(scaleanchor='y'),  # Ensures that x and y have the same scale
    coloraxis_colorbar=dict(title='Intensity')  # Color bar for intensity scale
)

# Step 4: Display the plot
fig.show()
```
