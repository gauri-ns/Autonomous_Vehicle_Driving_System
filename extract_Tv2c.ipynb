```python
import os
import csv

# Paths to the input and output directories
input_folder = 'E:/intern_work_testing/calib_orig'
output_folder = 'E:/intern_work_testing/calib'

# Create the output directory if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        # Full path to the current text file
        file_path = os.path.join(input_folder, filename)
        
        # Define the output CSV file path
        output_csv_path = os.path.join(output_folder, filename.replace('.txt', '.csv'))
        
        # Initialize a variable to hold the Tr_velo_to_cam matrix values
        tr_velo_to_cam_matrix = []

        # Read the calibration file
        with open(file_path, 'r') as file:
            for line in file:
                # Find the line that starts with "Tr_velo_to_cam"
                if line.startswith("Tr_velo_to_cam"):
                    # Extract the matrix values after the "Tr_velo_to_cam:" label
                    matrix_values = line.split()[1:]
                    # Convert the values to float and add to the matrix list
                    tr_velo_to_cam_matrix = [float(value) for value in matrix_values]
                    break

        # Reshape the list into a 3x4 matrix format
        tr_velo_to_cam_matrix = [tr_velo_to_cam_matrix[i:i+4] for i in range(0, len(tr_velo_to_cam_matrix), 4)]

        # Write the matrix to a CSV file
        with open(output_csv_path, 'w', newline='') as csv_file:
            writer = csv.writer(csv_file)
            # Write each row of the matrix to the CSV
            writer.writerows(tr_velo_to_cam_matrix)

        print(f"Tr_velo_to_cam matrix for {filename} has been saved to {output_csv_path}")

print("All files have been processed and saved to CSV.")
```
