```python
# edit line 8,9

import pandas as pd


for i in range(0,50):
# Load the CSV files
    labels_path = f'E:/intern_work_testing/label_obj/{i:06}.csv'  # Path to labels.csv
    pts_path = f'E:/intern_work_testing/2d_front_view/{i:06}.csv'        # Path to pts.csv

    labels_df = pd.read_csv(labels_path)
    pts_df = pd.read_csv(pts_path)

    # Define a function to crop points for a single object
    def crop_window(center_x, center_y, points_df, window_width=118, window_height=50):
        x_min = center_x - window_width // 2 
        x_max = center_x + window_width // 2 
        y_min = center_y - window_height // 2 
        y_max = center_y + window_height // 2 

        # Filter points within the window
        cropped_points = points_df[
            (points_df['Xfront'] > x_min) & (points_df['Xfront'] <= x_max) &
            (points_df['Yfront'] > y_min) & (points_df['Yfront'] <= y_max)
        ]
        return cropped_points[['Xfront', 'Yfront', 'intensity', 'time']].to_dict(orient='records')

    # Add a new column for cropped data
    labels_df['cropped_data'] = labels_df.apply(
        lambda row: crop_window(row['Xfront'], row['Yfront'], pts_df), axis=1
    )

    # Save back to the original file
    labels_df.to_csv(labels_path, index=False)

    print("Cropped data added to the labels file.")
```
