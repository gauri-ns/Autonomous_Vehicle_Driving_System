```python
import os
import csv

# Paths to the input and output directories
input_folder = 'E:/intern_work_testing/labels_orig'  # Replace with the path to your KITTI labels folder
output_folder = 'E:/intern_work_testing/labels'  # Replace with the path to your output CSV folder

# Create the output directory if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Define the header for the CSV files
header = [
    'ObjectType', 'Truncation', 'Occlusion', 'Alpha', 
    'Left', 'Top', 'Right', 'Bottom', 
    'Height', 'Width', 'Length', 
    'X', 'Y', 'Z', 'RotationY'
]

# Loop through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        # Full path to the current text file
        file_path = os.path.join(input_folder, filename)
        
        # Define the output CSV file path
        output_csv_path = os.path.join(output_folder, filename.replace('.txt', '.csv'))
        
        # Open the input text file and output CSV file
        with open(file_path, 'r') as txt_file, open(output_csv_path, 'w', newline='') as csv_file:
            writer = csv.writer(csv_file)
            
            # Write the header row
            writer.writerow(header)
            
            # Process each line in the text file
            for line in txt_file:
                # Split the line into individual values based on whitespace
                values = line.strip().split()
                
                # Write the values to the CSV file
                writer.writerow(values)

        print(f"Converted {filename} to {output_csv_path}")

print("All files have been processed and converted to CSV format.")
```
