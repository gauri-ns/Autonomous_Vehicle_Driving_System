```python
# edit line no. 11, 
import pandas as pd
import numpy as np
import os
import math

# Speed of light in air (m/s)
c = 2.99705 * 10**8  

# Folder containing CSV files
folder_path = 'E:/intern_work_testing/converted_csv_dataset/'  # Replace with the actual folder path

# Get list of CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Process each CSV file
for file in csv_files:
    # Load the CSV file
    file_path = os.path.join(folder_path, file)
    data = pd.read_csv(file_path)

    # Calculate the distance r = sqrt(x^2 + y^2 + z^2)
    data['r'] = np.sqrt(data['x']**2 + data['y']**2 + data['z']**2)
    
    # Calculate the time = 2 * r / c
    data['time'] = 2 * data['r'] / c
    
    # Drop the 'r' column if you only want to keep the 'time' column
    data = data.drop(columns=['r'])

    # Save the updated CSV file (you can overwrite the original or save a new file)
    updated_file_path = os.path.join(folder_path, f"{file}")
    data.to_csv(updated_file_path, index=False)

    print(f"Processed and saved: {updated_file_path}")
```
